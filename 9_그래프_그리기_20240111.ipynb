# 1. MatPlotlib

> ## 1.1 Matplotlib 이란 ?

* 주어진 데이터 이용하여 다양한 형태의 그래프를 그리는 함수 제공
* 데이터를 그래프로 표현 -> 데이터의 분포나 항목 간의 상관관계 쉽게 파악
* 'math+plot+library' or 'pip install matplotlib' 명령 실행하여 설치

> ## 1.2 Matplotlib 으로 그래프 그리기

- (1) 선 그래프 : plot()


```python
# plot(x값, y값, 선스타일문자열, marker='마커기호', label='그래프이름')
# : y 값은 필수, 나머지 생략 가능

import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [1,4,9,16]

plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_excel('data1.xlsx')

plt.plot(df['kor'])
plt.show()
```

```python
df = pd.read_excel('data1.xlsx')

plt.plot(df['kor'], 'k--', marker = 'o')
plt.show()         # 선스타일문자열 ; k-- & 마커기호 ; o
```

```python
plt.plot(df['kor'], 'k--', marker = 'o', label = 'korean')
plt.plot(df['math'], marker = '^', label = 'math')
plt.legend()

plt.show()
```

```python
plt.plot(df['kor'], 'r', marker = 'o', label = 'korean')
plt.plot(df['math'], 'b', marker = '^', label = 'math')
plt.legend()
plt.show()
```

- (2) 산점도 그래프 : scatter()

```python
# scatter(x값, y값, c='색문자', marker='마커기호', label='그래프이름')
# : c, marker, label 생략 가능

x = [6,2,1,1,4,4,6,4,7,7]
y = [5,4,2,9,9,0,10,3,1,2]

plt.scatter(x, y)
plt.show()
```

```python
# Mission 1

# Numpy를 이용하여 생성한 값에 대한 산점도 그래프 그리기
# x축 값과 y축 값을 numpy의 arange(), random.randint() 사용하여 생성
# 그것을 여러 형식의 산점도 그래프로
# x=np.arange(1,11)
# y=np.random.randint(1,30,10)

import numpy as np

x = np.arange(1,11)
y = np.random.randint(1,30,10)

plt.scatter(x, y, c='r', marker='^')
plt.show()
```

- (3) 막대 그래프 : bar()/barh()

```python
# bar(x값, y값, color='색 문자') : 세로 막대 그래프
# barh(x값, y값, color='색 문자') : 가로 막대 그래프
# : color 값 생략 가능

x = ['A','B','C']
y = [100, 150, 120]

plt.bar(x,y)
plt.show()
```

- (4) 빈도 그래프(히스토그램) : hist()

```python
# hist(x값, bins=x값 구간 개수, color='색 문자', align='left'/'mid'/'right'중 하나)
# : bins ; 데이터의 구간 개수 할당 (지정X - 10 할당)
# : align ; 막대가 x축 눈금 대표값을 기준으로 왼쪽/가운데/오른쪽으로 정렬

x= [1,1,1,2,3,2,4]

plt.hist(x, bins=4, color='green', align='mid')
plt.show()
```

- (5) 그래프 제목, x축 제목, y축 제목 지정 : title(), xlabel(), ylabel()

```python
# title('문자열') : 그래프의 제목
# xlabel('문자열') : x축의 제목
# ylabel('문자열') : y축의 제목

x = ['A','B','C']
y = [100, 150, 120]

plt.bar(x,y)
plt.title('Test Graph')
plt.xlabel('Group')
plt.ylabel('Score')
plt.show()
```

- (6) 한 화면에 여러 개 그래프 나타내기 : subplots()

```python
# subplots(행 개수, 열 개수, figsize=(가로길이, 세로길이))

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

fig, axs = plt.subplots(2,2)

axs[0,0].plot(x,y)  # plot() : 선그래프
axs[0,1].scatter(x,y) # scatter() : 산점도그래프
axs[1,0].bar(x,y) # bar() : 세로막대그래프
axs[1,1].barh(x,y)  # barh() : 가로막대그래프
fig.suptitle('Drawing Graph')
plt.show()
```

```python
# scatterplot(x='x축에 표시할 칼럼명', y='y축에 표시할 칼럼명',
#             hue='데이터 구분 칼럼명', data=데이터프레임)
# : hue값 지정 생략 가능

import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(x='petal_length',y='petal_width',hue='species',data=iris)
plt.title('Scatter Plot by seaborn', fontsize=20)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

iris = pd.read_csv('iris.csv')
iris.info()
```

```python
sns.scatterplot(x='petal_length',y='petal_width',hue='species',data=iris)
plt.title('Scatter Plot by seaborn', fontsize=20)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(x='species', data=iris)
plt.show()
```

```python
sns.barplot(x='species',y='petal_length',data=iris, errorbar='sd')
plt.show()
```

```python
sns.pairplot(iris, hue='species')
plt.show()
```

```python
sns.regplot(x='petal_length',y='petal_width', data=iris)
plt.show()
```

```python
df = iris.corr()
sns.heatmap(data=df, annot=True, fmt='.2f', cbar=False)
plt.show()
```
